```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load the Excel file and specify the sheet name
file_path = 'World Energy Overview.xlsx'
df = pd.read_excel(file_path, sheet_name='World Energy Overview')

# Ensure the 'Date' column is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Resample data annually by summing up values for each year
df_annual = df.resample('Y').sum()

# Plot observed annual data
ax = df_annual['Total Primary Energy Production'].plot(label='Observed', figsize=(14, 7))

# Forecast future values (adjust forecast steps to match annual data)
forecast_steps = 5  # Next 5 years

# Fit ARIMA model on annual data
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Define your p, d, q, P, D, Q, s parameters here based on your model selection criteria
p, d, q = 1, 1, 1
P, D, Q, s = 1, 1, 1, 12

model_arima = SARIMAX(df_annual['Total Primary Energy Production'], order=(p, d, q))
results_arima = model_arima.fit()

# Forecast with ARIMA
forecast_arima = results_arima.get_forecast(steps=forecast_steps)
forecast_arima_ci = forecast_arima.conf_int()

# Fit SARIMA model on annual data
model_sarima = SARIMAX(df_annual['Total Primary Energy Production'], 
                       order=(p, d, q), 
                       seasonal_order=(P, D, Q, s))
results_sarima = model_sarima.fit()

# Forecast with SARIMA
forecast_sarima = results_sarima.get_forecast(steps=forecast_steps)
forecast_sarima_ci = forecast_sarima.conf_int()

# Combine forecast index with the end of the observed data index
extended_index_arima = df_annual.index.union(pd.to_datetime(forecast_arima.predicted_mean.index))
extended_index_sarima = df_annual.index.union(pd.to_datetime(forecast_sarima.predicted_mean.index))

# Reindex forecast series and confidence intervals
forecast_arima_mean = forecast_arima.predicted_mean.reindex(extended_index_arima)
forecast_arima_ci = forecast_arima_ci.reindex(extended_index_arima)

forecast_sarima_mean = forecast_sarima.predicted_mean.reindex(extended_index_sarima)
forecast_sarima_ci = forecast_sarima_ci.reindex(extended_index_sarima)

# Plot ARIMA forecast
forecast_arima_mean.plot(ax=ax, label='ARIMA Forecast', alpha=.7)

# Plot SARIMA forecast
forecast_sarima_mean.plot(ax=ax, label='SARIMA Forecast', alpha=.7)

# Fill between confidence intervals for ARIMA
ax.fill_between(forecast_arima_ci.index,
                forecast_arima_ci.iloc[:, 0],
                forecast_arima_ci.iloc[:, 1], color='k', alpha=.2)

# Fill between confidence intervals for SARIMA
ax.fill_between(forecast_sarima_ci.index,
                forecast_sarima_ci.iloc[:, 0],
                forecast_sarima_ci.iloc[:, 1], color='b', alpha=.2)

# Set labels
ax.set_xlabel('Year')
ax.set_ylabel('Total Primary Energy Production')

# Format x-axis to show years only
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # Display only years
ax.xaxis.set_major_locator(mdates.YearLocator())          # Set major ticks at yearly intervals

# Set x-axis limits to include all observed and forecasted years
ax.set_xlim([df_annual.index[0], forecast_arima_ci.index[-1]])

# Rotate date labels for better readability
plt.xticks(rotation=45)

# Add legend
plt.legend()

# Show plot
plt.show()
```
